# Como crear un entorno virtual

## Windows

 instalar virtualenv utilizando el siguiente comando en PowerShell:

pip install virtualenv

### Utiliza los siguientes comandos para navegar al directorio y crear el entorno virtual:

cd ruta\al\tu\nuevo\directorio
virtualenv nombre_del_entorno

### Para Activarlo

.\nombre_del_entorno\Scripts\Activate


### Para desactivarlo

deactivate


Posibles errores 

No habilita el ambiente

Puedes habilitar la ejecución de scripts temporalmente ejecutando el siguiente comando en PowerShell con privilegios de administrador:

Set-ExecutionPolicy RemoteSigned -Scope Process

Una vez que hayas terminado de trabajar con el entorno virtual, puedes restaurar la política de ejecución a su valor original ejecutando el siguiente comando:

Set-ExecutionPolicy -Default -Scope Process
